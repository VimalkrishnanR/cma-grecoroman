# Cleveland Museum of Art API 
### Interpretation of the Greek and Roman Art Department Collections

```python
#importing python packages
import requests
url = "https://openaccess-api.clevelandart.org/api/artworks/?department=Greek%20and%20Roman%20Art" 
response = requests.get(url) 
data = response.json()
import pandas as pd   
```

### Composition of Artworks by 'type' and 'technique'

```python
data.keys()
```

```python
#Assigning json data to a variable named 'list_of_artworks'. 
list_of_artworks = data['data']
```

```python
#loading the json data into a dataframe so that I can work with the imported pandas package. 
df = pd.DataFrame(list_of_artworks)
```

```python
#Making a list named relevant_columns to extract the columns needed for basic analyses: 
#I am then assigning the extracted columns to a mini data frame named 'artworks.'
#To find the composition of artworks by technique and type. 
relevant_columns = ['id', 'title', 'technique', 'type', 'did_you_know', 'description']
artworks = df[relevant_columns]
artworks.head()
```

```python
artworks.shape
```

```python
artworks.columns
```

```python
#Counting artworks by type.
count_of_types = artworks['type'].value_counts()
print("--- Different types of artoworks and the number of pieces for each type ---")
count_of_types
```

```python
#to be epxlored later 
forgery_type = artworks[artworks['type'] == 'Forgery']
forgery_type
```

```python
#Counting artworks by technique/medium. Note: technique refers to materials of creation. 'Medium' is an alias for technique. 
#In some instances, type and technique could be the same. For e.g. 'Ceramic' denotes 'type' and 'ceramic' denotes 'technique.'
#Source: Cleveland Museum of Art API documentation: https://openaccess-api.clevelandart.org/
count_of_techniques = artworks['technique'].value_counts()
print("----- Counts of medium of artworks ------")
count_of_techniques
```

```python
#Here, I am finding the top 5 artworks by 'type' and 'technique'. This data lends itself well to a pictogram bar chart. 
#I am also saving the top 5 'types' and 'techniques' as a csv so that I can work on flourish to make a pictogram.  
top_5_types = artworks['type'].value_counts().head(5)
top_5_techniques = artworks['technique'].value_counts().head(5)
top_5_types.to_csv('cleveland_top_types.csv')
top_5_techniques.to_csv('cleveland_top_techniques.csv')
                          
```

```python
#Here, I want to connect 'type' and 'technique' to answer questions like how many 'Coins' are made out of 'silver?'
# This is for a SANKEY diagram. 
# I am making two lists to hold top 5 values under 'type' column and 'technique' column. 
# Then I am querying my mini dataframe named 'artworks'to find instances that match elements in the 'top_types' and 'top_techniques' lists.
# Then, I am assigning the matched instances to a subset dataframe named 'selected_artworks.' 
# The 'selected_artworks dataframe' (a subset of the 'artworks' dataframe) contains only data related to top 5 types and top 5 techniques.
types_top_5 = ['Sculpture', 'Coins', 'Ceramic', 'Jewelry', 'Metalwork']
techniques_top_5 = ['silver', 'ceramic', 'bronze', 'marble', 'terracotta'] 
selected_artworks = artworks.query("type in @types_top_5 and technique in @techniques_top_5")
print(len(selected_artworks))
type(selected_artworks)
```

```python
#Here I am cross tabulating the data by counting the 'type' of artwork and then counting the 'technique' to find pair counts.
#https://pandas.pydata.org/docs/reference/api/pandas.crosstab.html
sankey_grid = pd.crosstab(selected_artworks['type'], selected_artworks['technique'])
print (sankey_grid)
```

```python
#Here I am parsing the selected_artworks dataframe (a subset of the 'artworks'dataframe) to count the number of pairs of type and techniqe. 
#For e.g., Coins -silver, Sculpture - marble. I am assigning the count of these pairs (.value_counts) to 'Value'(reset_index). 
# I am adding 'Value' as a column to the dataframe 'sankey_counts' (a subset of the selected_artworks dataframe) as 
#the selected_artworks df does not have 'Values' as a column listing the counts of the type-techniqe pairs
sankey_counts = selected_artworks[['type','technique']].value_counts().reset_index(name = 'Value')
sankey_counts.columns = ['source', 'Target', 'Value'] #flourish accepts these three column headers for sankey charts.
#pair_counts.head(10)
sankey_counts.to_csv("sankey_type_technique.csv", index = False)
```
